## Just the not overlapping data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Step 1: Combine all light curves into a single DataFrame
all_data = []
for lc in data_all:
    df = pd.DataFrame({'time': lc.time.value, 'flux': lc.flux, 'sector': lc.sector})
    all_data.append(df)

# Merge all sectors into one DataFrame
df_all = pd.concat(all_data).sort_values('time')

# Step 2: Count how many sectors have data at each time point
df_all['count'] = df_all.groupby('time')['time'].transform('count')

# Step 3: Keep only data points where count = 1 (i.e., no overlap)
df_filtered = df_all[df_all['count'] == 1]

# Step 4: Plot the non-overlapping data
plt.figure(figsize=(12, 6))

for sector in df_filtered['sector'].unique():
    sector_data = df_filtered[df_filtered['sector'] == sector]
    plt.plot(sector_data['time'], sector_data['flux'], label=f"Sector {sector}", linewidth=0, marker='.',alpha=0.7)
plt.xlabel("Time [BTJD days]")
plt.ylabel("Flux [e s]")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ensure time is a numeric array
time_values = data.time.value if hasattr(data.time, "value") else data.time

# Normalize the flux using median
med = np.nanmedian(data.flux)
sec_norm = data.flux / med

# Convert to NumPy array if it's a masked quantity
flux_values = sec_norm.value if hasattr(sec_norm, "value") else sec_norm

# Plot
plt.plot(time_values, flux_values, linestyle="None", marker=".", alpha=0.4)

plt.xlabel("Time [BTJD days]")
plt.ylabel("Normalized Flux")
plt.show()
```

## BINNING ON WHOLE GRAPH(bin=50 days)

```python
bin_size = 50  #how many points average taken
time_values = data.time.value if hasattr(data.time, "value") else np.array(data.time)

# Convert flux to NumPy array, handling MaskedNDArray
flux_values = sec_norm.filled(np.nan) if hasattr(sec_norm, "filled") else np.array(sec_norm)

# Sort data by time (just in case)
sorted_indices = np.argsort(time_values)
time_values = time_values[sorted_indices]
flux_values = flux_values[sorted_indices]

# Bin the data
binned_time = np.array([np.mean(time_values[i : i + bin_size]) for i in range(0, len(time_values) - bin_size, bin_size)])
binned_flux = np.array([np.median(flux_values[i : i + bin_size]) for i in range(0, len(flux_values) - bin_size, bin_size)])

# Remove NaNs safely
binned_flux = np.nan_to_num(binned_flux, nan=np.nanmedian(flux_values))

# Plot original and binned data
plt.figure(figsize=(10, 5))
plt.plot(time_values, flux_values, 'k.', alpha=0.3, label="Original Data")
plt.plot(binned_time, binned_flux, 'r-', marker='o', markersize=4, label=f"Binned Data (bin size={bin_size})")

# Zoom in for better visualization

plt.xlabel("Time [BTJD days]")
plt.ylabel("Normalized Flux")
plt.legend()
plt.show()
```
